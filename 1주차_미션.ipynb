````python
import pandas as pd
# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df

# 2011년도 이후 데이터만 모으기
df1 = df[df['Year'] >= 2011]

# pivot_table 기능을 활용하여 형태 변환하기
result = df1.pivot_table(index = 'Year', columns = 'Country', values = 'Life_Expectancy')
result
````
